Skewness is a measure of the asymmetry of a probability distribution around its mean.

**Types of skewness**

Positive Skewness (Right-Skewed): The right tail of the distribution is longer or fatter than the left one. This means the data has a few very high values that pull the mean to the right.

Negative Skewness (Left-Skewed): The left tail of the distribution is longer or fatter than the right one. This indicates a few very low values that pull the mean to the left.

Zero Skewness: The distribution is perfectly symmetrical, like a normal or bell-shaped curve.

```python
# Function to remove outliers in one column using IQR
def remove_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    # Keep only rows where column is within bounds
    return data[(data[column] >= lower) & (data[column] <= upper)]

# Choose numeric columns you want to clean
numeric_cols = data.select_dtypes(include=np.number).columns.tolist()

# Remove outliers for each column iteratively
data_clean = data.copy()
for col in numeric_cols:
    data_clean = remove_outliers_iqr(data_clean, col)

# Now df_clean is the dataset after outlier removal
print("Before count:", len(data))
print("After count:", len(data_clean))
print("\nStatistics after removal:")
print(data_clean.describe())
```

```python
#import necessary libraries
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVC
```

```python
#Initialize the regression
Regression={
    "DecisionTreeRegressor":DecisionTreeRegressor(random_state=42),
    "RandomForestRegressor":RandomForestRegressor(random_state=42),
    }
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
#Train,evaluate and store results
from sklearn.metrics import mean_squared_error, r2_score

#initialize an empty dictionary to store the results
results= {}

#Train and evaluate each regression
for name, regressor in Regression.items():
  # Drop rows with NaN values in the target variable for training
  y_train_cleaned = y_train.dropna()
  x_train_scaled_cleaned = x_train_scaled.loc[y_train_cleaned.index]

  #Train the model
  regressor.fit(x_train_scaled_cleaned,y_train_cleaned)
  #Make Predictions
  y_pred=regressor.predict(x_test_scaled)

  # Drop rows with NaN values in y_test and corresponding y_pred
  y_test_cleaned = y_test.dropna()
  y_pred_cleaned = pd.Series(y_pred, index=x_test_scaled.index).loc[y_test_cleaned.index]

  #calculate metrics
  mse=mean_squared_error(y_test_cleaned,y_pred_cleaned)
  r2=r2_score(y_test_cleaned,y_pred_cleaned)
  #store the results
  results[name]={'MSE':mse,'R-squared':r2}

  #convert the results dictionary to a Dataframe
  results_df=pd.DataFrame(results).T.sort_values(by='R-squared', ascending=False)

  #print the DateFrame
  print("model performance comparison:\n")
  print(results_df)

  #select best model based on R-squared score
  best_model_name=results_df['R-squared'].idxmax()
  print(f"\nBest model for regression (R-squared prioritized):{best_model_name}")
```

# **ANOVA**
ANOVA stands for Analysis of Variance.
It’s a statistical test used to compare the means (averages) of three or more groups to see if they are significantly different from each other.

```python
# Check normality (Shapiro test)
from scipy import stats
for make in data['make'].unique():
    # Use 'sale_price' as the target variable
    stat, p = stats.shapiro(data.loc[data['make'] == make, 'sale_price'])
    print(f"{make}: p={p:.3f}")
```

**One** **Way** **ANNOVA**
# **INSIGHTS**
If p < 0.05, there is a significant difference in mean price among at least one of the groups.

```python
# One-way ANOVA for price across brands
groups = [data.loc[data['make'] == b, 'sale_price'].dropna() for b in data['make'].unique()]
f_stat, p_value = stats.f_oneway(*groups)

print(f"F-statistic: {f_stat:.3f}, p-value: {p_value:.3f}")
```

Hyperparameter tuning means finding the best set of hyperparameters (the “settings” of your ML model) that make your model perform as well as possible on unseen data.

```python
# Define the parameter grid to search
param_grid = {
    'n_estimators': [50, 100],
    'max_depth': [10, 20],
    'min_samples_split': [2, 5],
    'min_samples_leaf': [1, 2]
}

# Initialize GridSearchCV
grid_search = GridSearchCV(estimator=rf_model, param_grid=param_grid, cv=3, n_jobs=-1, verbose=2)

# Fit the grid search to the data
grid_search.fit(X_train, y_train)

# Get the best parameters and best score
best_params = grid_search.best_params_
best_score = grid_search.best_score_

print(f"Best parameters found: {best_params}")
print(f"Best cross-validation score: {best_score:.4f}")
```

```python

# Build the final model with the best parameters
final_rf_model = RandomForestRegressor(**best_params, random_state=42)
final_rf_model.fit(X_train, y_train)

# Evaluate the final model on the test set
final_preds = final_rf_model.predict(X_test)
final_r2 = r2_score(y_test, final_preds)
final_mse = mean_squared_error(y_test, final_preds)

print(f"Final Tuned Random Forest R-squared: {final_r2:.4f}, MSE: {final_mse:.2f}")
```

# **ROC**

1. ROC Curve:

This is a graph showing the performance of a classification model at all classification thresholds. It plots the True Positive Rate (TPR) against the False Positive Rate (FPR) at various threshold settings.
TPR (Sensitivity): The proportion of actual positive cases that are correctly identified.

- FPR (1 - Specificity):

- The proportion of actual negative cases that are incorrectly identified as positive.
2.AUC:

The Area Under the ROC Curve. It provides a single scalar value that summarizes the model's ability to distinguish between the positive and negative classes. An AUC of 1.0 represents a perfect classifier, while an AUC of 0.5 indicates a model that performs no better than random chance.


```python

from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt

# Convert the regression problem to a binary classification problem
# We'll classify a car as "expensive" if its price is above the median price
median_price = np.median(y_train)
y_train_binary = (y_train > median_price).astype(int)
y_test_binary = (y_test > median_price).astype(int)

# Use a classifier model for the binary problem
from xgboost import XGBClassifier
clf_model = XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42)
clf_model.fit(X_train, y_train_binary)

# Get predicted probabilities for the positive class
y_pred_proba = clf_model.predict_proba(X_test)[:, 1]

# Calculate ROC curve and AUC
fpr, tpr, thresholds = roc_curve(y_test_binary, y_pred_proba)
roc_auc = auc(fpr, tpr)

# Plot the ROC curve
plt.figure()
plt.plot(fpr, tpr, color='darkorange', lw=2, label='ROC curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc="lower right")
plt.show()
```
